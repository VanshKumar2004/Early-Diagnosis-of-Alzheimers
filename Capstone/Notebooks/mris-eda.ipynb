## Loading Images By The Classes


The very first step to start eda for images datasets is to load all the images to later use for different things like plotting. Since we have our dataset already split up into train and test so to prepare and preprocess the data we can use training data. We have data distributed in 4 classes:

- `Mild Demented`

- `Moderate Demented`

- `Non Demented`

- `Very Mild Demented`

we will start by creating a list of photo paths for each class:


```python
# Using glob to store the paths in multiple lists
mild_files = glob('../Alzheimer_s Dataset/train/MildDemented/*.jpg')
moderate_files = glob('../Alzheimer_s Dataset/train/ModerateDemented/*.jpg')
non_demented_files = glob('../Alzheimer_s Dataset/train/NonDemented/*.jpg')
very_mild_files = glob('../Alzheimer_s Dataset/train/VeryMildDemented/*.jpg')
```

Now Since we have loaded the paths to 4 different lists now we will make new 4 lists(one for each class) which contain the images itself.
Lets start by loading the MildDemented images

```python
img_mild = []
#using for loop to load multiple lists
for image_path in mild_files:
    img_mild.append(cv2.imread(image_path)) #cv2 helps in reading the images
```

For our sanity check we will print the length of the images list and shape of any image

```python
#Sanity Check
print(len(img_mild), img_mild[0].shape)
```

Now lets Load the ModerateDemented images:

```python
img_moderate = []
#using for loop to load multiple lists
for image_path in moderate_files:
    img_moderate.append(cv2.imread(image_path)) #cv2 helps in reading the images

```

```python
#sanity check
print(len(img_moderate), img_moderate[0].shape)
```

lets Load the NonDemented images:

```python
img_non = []
#using for loop to load multiple lists
for image_path in non_demented_files:
    img_non.append(cv2.imread(image_path)) #cv2 helps in reading the images

```

```python
#Sanity Check
print(len(img_non), img_non[0].shape)
```

lets Load the VeryMildDemented images:

```python
img_very_mild = []
#using for loop to load multiple lists
for image_path in very_mild_files:
    img_very_mild.append(cv2.imread(image_path)) #cv2 helps in reading the images

```

```python
#Sanity Check
print(len(img_very_mild),img_very_mild[0].shape)
```

## Image Channels

### RGB Channels:

An RGB image has three channels: red, green, and blue. RGB channels roughly follow the color receptors in the human eye, and are used in computer displays and image scanners.

Source: Wikipedia#:~:text=RGB%20images,-An%20RGB%20image&text=If%20the%20RGB%20image%20is,intensities%20between%200%20and%20255.)


Since we previously analysed that our each image has a shape of (208, 176, 3) which indicates the 3 RGB Channels in our each image, so now lets take one image from each class and look at them in 3 different channels individually:

Lets start with a `MildDemented` MRI:

```python
plt.subplots(1, 3, figsize=(15, 5))

for i in range(0,3):
    plt.subplot(1,3,i+1)
    plt.imshow(img_mild[1][:,:,i])
    plt.title(f'Image {i + 1}')
    plt.axis("off")
   
   
```

Now as a confirmation on our image color channels requirement we will plot histograms with individual color channels. This will later help us to decide if we wanna eliminate any certain channel. To execute this idea we will take one image(MRI) from each class and plot a histogram with 3 different channels 

```python
# using cv2 to calculate hist
hist1 = cv2.calcHist([img_mild[1]],[0],None,[256],[0,256])
hist2 = cv2.calcHist([img_mild[1]],[1],None,[256],[0,256])
hist3 = cv2.calcHist([img_mild[1]],[2],None,[256],[0,256])

plt.figure()

plt.plot(hist1, label='Channel 0')
plt.plot(hist2, label='Channel 1')
plt.plot(hist3, label='Channel 2')
plt.title("Distribution of Mild Demented MRI with different color channels\n".title())
plt.legend()  # This will display the legend with the provided labels
plt.show()
```

The overlapping of distribution as seen above proves the point that MRI images typically have a grayscale appearance even if they are stored in a three-channel format, Now lets see if this is true for the remaining classes as well:

Now since we checked distribution of each MRI from each class we can confirm that keeping a single color channel would be beneficial in terms of efficiency in the future:

Lets convert all the images in each class to a single color channel and we acheive this by using cv2.COLOR_BGR2GRAY

Lets start with `MildDemented` class:


```python
# using a for loop to convert all pictures to grayscale in this class
for i in range(len(img_mild)):
    img_mild[i] = cv2.cvtColor(img_mild[i], cv2.COLOR_BGR2GRAY) 
```

Previously in our EDA process we visualised some histograms with showed pixel distributions of a single image in each category with different color channels. Now since we have set one color channel so we can take the averages of images in each class and display the overall distribution of image pixels for each class:

Lets start by calculating the averages for `MildDemented` class:


```python
# using a for loop to calculate the average of each image in my list of images and storing it another list of averages
averages_mild = []

for i in range(len(img_mild)):
    averages_mild.append(img_mild[i].mean())
```

```python
#sanity check
len(averages_mild) == len(img_mild)
```

Now lets plot these averages on a histogram to see the overall class pixel distribution

```python
plt.figure()

plt.hist(averages_mild)
plt.title("Average distribution of pixels for MildDemented MRIs ")
plt.show()
```
